**Chapter 1: Probability Theory**

```python
# Conceptos básicos
sample_space = "Conjunto de todos los resultados posibles de un experimento aleatorio."
event = "Subconjunto del espacio muestral."
probability = "Medida que asigna un valor entre 0 y 1 a un evento."

print("Conceptos básicos:")
print(f"Espacio muestral: {sample_space}")
print(f"Evento: {event}")
print(f"Probabilidad: {probability}\n")

# Operaciones con eventos
union_description = "La unión de A y B (A ∪ B) contiene todos los resultados que están en A o en B o en ambos."
intersection_description = "La intersección de A y B (A ∩ B) contiene los resultados que están en A y en B simultáneamente."
complement_description = "El complemento de A (Aᶜ) contiene todos los resultados que NO están en A."
independence_description = "A y B son independientes si la ocurrencia de uno no afecta la probabilidad del otro."
bayes_theorem_description = "El Teorema de Bayes actualiza la probabilidad de un evento dado que otro ya ocurrió."

print("Operaciones con eventos:")
print(f"Unión: {union_description}")
print(f"Intersección: {intersection_description}")
print(f"Complemento: {complement_description}")
print(f"Independencia: {independence_description}")
print(f"Teorema de Bayes: {bayes_theorem_description}\n")

# Reglas básicas de probabilidad
def probability_union(p_a, p_b, p_a_and_b):
    """Calcula la probabilidad de la unión de dos eventos."""
    return p_a + p_b - p_a_and_b

def probability_intersection(p_a_given_b, p_b):
    """Calcula la probabilidad de la intersección usando P(A|B) * P(B)."""
    return p_a_given_b * p_b

def probability_complement(p_a):
    """Calcula la probabilidad del complemento de un evento."""
    return 1 - p_a

# Independencia y probabilidad condicional
def conditional_probability(p_a_and_b, p_b):
    """Calcula la probabilidad condicional P(A|B)."""
    return p_a_and_b / p_b

def are_independent(p_a_and_b, p_a, p_b):
    """Verifica si dos eventos son independientes."""
    return p_a_and_b == p_a * p_b

# Teorema de Bayes
def bayes_theorem(p_b_given_a, p_a, p_b):
    """Calcula P(A|B) usando el teorema de Bayes."""
    return (p_b_given_a * p_a) / p_b

# Variables aleatorias
random_variable = "Función que asigna un valor numérico a cada resultado en el espacio muestral."
print("Variable aleatoria:")
print(f"{random_variable}\n")

# Distribución Bernoulli
bernoulli_distribution = "Modelo de un experimento de dos resultados: éxito (1) o fracaso (0)."
print("Distribución Bernoulli:")
print(f"{bernoulli_distribution}\n")

print("Capítulo 1: Teoría de la Probabilidad\n")


# Ejemplos de uso de funciones
print("Ejemplos de funciones:\n")

# Ejemplo: Unión de dos eventos
p_a = 0.3
p_b = 0.5
p_a_and_b = 0.1
print(f"Probabilidad de A o B (Unión): {probability_union(p_a, p_b, p_a_and_b)}")

# Ejemplo: Intersección de dos eventos
p_a_given_b = 0.4
p_b = 0.5
print(f"Probabilidad de A y B (Intersección): {probability_intersection(p_a_given_b, p_b)}")

# Ejemplo: Complemento de un evento
p_a = 0.7
print(f"Complemento de A: {probability_complement(p_a)}")

# Ejemplo: Probabilidad condicional
p_a_and_b = 0.2
p_b = 0.5
print(f"Probabilidad condicional P(A|B): {conditional_probability(p_a_and_b, p_b)}")

# Ejemplo: Eventos independientes
p_a = 0.5
p_b = 0.4
p_a_and_b = 0.2
print(f"¿Son A y B independientes?: {are_independent(p_a_and_b, p_a, p_b)}")

# Ejemplo: Teorema de Bayes
p_b_given_a = 0.8
p_a = 0.3
p_b = 0.5
print(f"P(A|B) usando Teorema de Bayes: {bayes_theorem(p_b_given_a, p_a, p_b)}")
```

**Chapter 2: Random Variables**

```python
# Conceptos básicos
random_variable_description = "Una variable aleatoria asigna valores numéricos a los resultados de un experimento aleatorio."
discrete_variable_description = "Variable aleatoria discreta: puede tomar valores contables, como el número de caras al lanzar una moneda varias veces."
continuous_variable_description = "Variable aleatoria continua: puede tomar infinitos valores dentro de un intervalo, como el tiempo o la distancia."

print("Variables Aleatorias:")
print(f"Definición: {random_variable_description}")
print(f"Variable Discreta: {discrete_variable_description}")
print(f"Variable Continua: {continuous_variable_description}\n")

# Funciones importantes
pmf_description = "PMF (Probability Mass Function): asigna una probabilidad a cada valor específico de una variable discreta."
pdf_description = "PDF (Probability Density Function): describe la densidad de probabilidad en variables continuas; la probabilidad real se calcula como un área bajo la curva."
cdf_description = "CDF (Cumulative Distribution Function): representa la probabilidad de que una variable aleatoria sea menor o igual a un valor dado."

print("Funciones asociadas:")
print(f"PMF: {pmf_description}")
print(f"PDF: {pdf_description}")
print(f"CDF: {cdf_description}\n")

# Definiciones adicionales
bernoulli_description = "Distribución Bernoulli: modela un experimento con dos resultados, éxito (1) o fracaso (0)."
binomial_description = "Distribución Binomial: número de éxitos en n pruebas independientes de Bernoulli."
poisson_description = "Distribución de Poisson: número de eventos que ocurren en un intervalo de tiempo o espacio fijo."

print("Distribuciones específicas:")
print(f"Bernoulli: {bernoulli_description}")
print(f"Binomial: {binomial_description}")
print(f"Poisson: {poisson_description}\n")

# Definiciones matemáticas
def expected_value(values, probabilities):
    """Calcula el valor esperado E[X] para una variable aleatoria discreta."""
    return sum(v * p for v, p in zip(values, probabilities))

def variance(values, probabilities):
    """Calcula la varianza Var(X) para una variable aleatoria discreta."""
    mean = expected_value(values, probabilities)
    return sum(p * (v - mean)**2 for v, p in zip(values, probabilities))

def std_deviation(values, probabilities):
    """Calcula la desviación estándar de X."""
    return variance(values, probabilities) ** 0.5

print("Propiedades matemáticas:")
print("Valor Esperado (E[X]): suma de cada valor multiplicado por su probabilidad.")
print("Varianza (Var[X]): esperanza del cuadrado de la diferencia respecto al valor esperado.")
print("Desviación estándar: raíz cuadrada de la varianza.\n")

print("Capítulo 2: Variables Aleatorias.\n")

# Ejemplos de uso de funciones

print("Ejemplos de funciones:\n")

# Ejemplo: Variable aleatoria discreta - Lanzamiento de un dado justo
values = [1, 2, 3, 4, 5, 6]
probabilities = [1/6] * 6

print("Variable: Dado justo")
print(f"Valor esperado del dado: {expected_value(values, probabilities)}")
print(f"Varianza del dado: {variance(values, probabilities)}")
print(f"Desviación estándar del dado: {std_deviation(values, probabilities)}\n")

# Ejemplo: Distribución Bernoulli
print("Distribución Bernoulli (p=0.3):")
bernoulli_values = [0, 1]
bernoulli_probabilities = [0.7, 0.3]
print(f"Valor esperado: {expected_value(bernoulli_values, bernoulli_probabilities)}")
print(f"Varianza: {variance(bernoulli_values, bernoulli_probabilities)}\n")

# Ejemplo: Distribución Binomial (n=5, p=0.5)
from scipy.stats import binom

n = 5
p = 0.5
x = [0, 1, 2, 3, 4, 5]
binomial_probabilities = [binom.pmf(k, n, p) for k in x]
print("Distribución Binomial (n=5, p=0.5):")
print(f"Valores posibles: {x}")
print(f"Probabilidades: {binomial_probabilities}")
print(f"Valor esperado: {expected_value(x, binomial_probabilities)}")
print(f"Varianza: {variance(x, binomial_probabilities)}\n")

# Ejemplo: Distribución de Poisson (λ=3)
from scipy.stats import poisson

lambda_poisson = 3
x_poisson = [0, 1, 2, 3, 4, 5]
poisson_probabilities = [poisson.pmf(k, lambda_poisson) for k in x_poisson]
print("Distribución de Poisson (λ=3):")
print(f"Valores posibles: {x_poisson}")
print(f"Probabilidades: {poisson_probabilities}")
print(f"Valor esperado aproximado: {expected_value(x_poisson, poisson_probabilities)}")
print(f"Varianza aproximada: {variance(x_poisson, poisson_probabilities)}")
```

**Chapter 3: Discrete Probability Distributions**

```python
# Conceptos básicos
print("Distribuciones Discretas de Probabilidad:\n")

binomial_desc = "Distribución Binomial: cuenta el número de éxitos en n ensayos independientes con probabilidad p."
geometric_desc = "Distribución Geométrica: modela el número de ensayos hasta el primer éxito."
negative_binomial_desc = "Distribución Binomial Negativa: modela el número de ensayos hasta obtener r éxitos."
hypergeometric_desc = "Distribución Hipergeométrica: mide el número de éxitos en una muestra sin reemplazo."

print(f"Binomial: {binomial_desc}")
print(f"Geométrica: {geometric_desc}")
print(f"Binomial Negativa: {negative_binomial_desc}")
print(f"Hipergeométrica: {hypergeometric_desc}\n")

# Importar librerías
from scipy.stats import binom, geom, nbinom, hypergeom
import matplotlib.pyplot as plt
import numpy as np

# Definiciones matemáticas generales
def expected_value(values, probabilities):
    """Calcula el valor esperado E[X]."""
    return sum(v * p for v, p in zip(values, probabilities))

def variance(values, probabilities):
    """Calcula la varianza Var(X)."""
    mean = expected_value(values, probabilities)
    return sum(p * (v - mean)**2 for v, p in zip(values, probabilities))

# Ejemplos de cada distribución

# 1. Binomial (n=10, p=0.5)
print("Distribución Binomial (n=10, p=0.5):")
n, p = 10, 0.5
x = np.arange(0, n+1)
binomial_probs = binom.pmf(x, n, p)
print(f"Valores posibles: {x}")
print(f"Probabilidades: {binomial_probs}\n")

# Grafica
plt.bar(x, binomial_probs)
plt.title('Distribución Binomial (n=10, p=0.5)')
plt.xlabel('Número de éxitos')
plt.ylabel('Probabilidad')
plt.show()

# 2. Geométrica (p=0.3)
print("Distribución Geométrica (p=0.3):")
p_geom = 0.3
x_geom = np.arange(1, 11)
geom_probs = geom.pmf(x_geom, p_geom)
print(f"Valores posibles: {x_geom}")
print(f"Probabilidades: {geom_probs}\n")

# Grafica
plt.bar(x_geom, geom_probs)
plt.title('Distribución Geométrica (p=0.3)')
plt.xlabel('Número de ensayos hasta primer éxito')
plt.ylabel('Probabilidad')
plt.show()

# 3. Binomial Negativa (r=5, p=0.4)
print("Distribución Binomial Negativa (r=5, p=0.4):")
r, p_nb = 5, 0.4
x_nb = np.arange(5, 15)
nbinom_probs = nbinom.pmf(x_nb - r, r, p_nb)
print(f"Valores posibles: {x_nb}")
print(f"Probabilidades: {nbinom_probs}\n")

# Grafica
plt.bar(x_nb, nbinom_probs)
plt.title('Distribución Binomial Negativa (r=5, p=0.4)')
plt.xlabel('Total de ensayos hasta 5 éxitos')
plt.ylabel('Probabilidad')
plt.show()

# 4. Hipergeométrica
print("Distribución Hipergeométrica (N=50, K=10, n=5):")
N, K, n_hg = 50, 10, 5
x_hg = np.arange(0, n_hg+1)
hypergeom_probs = hypergeom.pmf(x_hg, N, K, n_hg)
print(f"Valores posibles: {x_hg}")
print(f"Probabilidades: {hypergeom_probs}\n")

# Grafica
plt.bar(x_hg, hypergeom_probs)
plt.title('Distribución Hipergeométrica (N=50, K=10, n=5)')
plt.xlabel('Número de éxitos en muestra')
plt.ylabel('Probabilidad')
plt.show()
```

**Chapter 4: Comtimuous Probability Distributions**

```python
# Conceptos básicos
print("Distribuciones Continuas de Probabilidad:\n")

pdf_desc = "PDF (Probability Density Function): función que describe la probabilidad relativa de una variable continua en un valor específico."
cdf_desc = "CDF (Cumulative Distribution Function): probabilidad de que una variable continua sea menor o igual a un valor dado."

uniform_desc = "Distribución Uniforme Continua: todos los valores en un intervalo tienen la misma probabilidad."
normal_desc = "Distribución Normal: distribución simétrica en forma de campana, centrada en su media."
exponential_desc = "Distribución Exponencial: modela el tiempo entre eventos en un proceso de Poisson."

print(f"PDF: {pdf_desc}")
print(f"CDF: {cdf_desc}\n")
print(f"Distribución Uniforme: {uniform_desc}")
print(f"Distribución Normal: {normal_desc}")
print(f"Distribución Exponencial: {exponential_desc}\n")

# Importar librerías
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform, norm, expon

# Definiciones
def expected_value_continuous(pdf, a, b):
    """Calcula el valor esperado E[X] de una variable continua entre [a,b]."""
    from scipy.integrate import quad
    return quad(lambda x: x * pdf(x), a, b)[0]

def variance_continuous(pdf, mean, a, b):
    """Calcula la varianza Var(X) de una variable continua entre [a,b]."""
    from scipy.integrate import quad
    return quad(lambda x: (x - mean)**2 * pdf(x), a, b)[0]

# Ejemplos de distribuciones

# 1. Distribución Uniforme continua en [0, 1]
print("Distribución Uniforme continua en [0,1]:")
a, b = 0, 1
x_uniform = np.linspace(a, b, 100)
pdf_uniform = uniform.pdf(x_uniform, loc=a, scale=b-a)
cdf_uniform = uniform.cdf(x_uniform, loc=a, scale=b-a)

# Grafica
plt.plot(x_uniform, pdf_uniform, label="PDF")
plt.plot(x_uniform, cdf_uniform, label="CDF", linestyle='dashed')
plt.title('Distribución Uniforme Continua [0,1]')
plt.xlabel('x')
plt.ylabel('Densidad / Probabilidad')
plt.legend()
plt.show()

# 2. Distribución Normal estándar (media=0, sigma=1)
print("Distribución Normal estándar (μ=0, σ=1):")
mu, sigma = 0, 1
x_normal = np.linspace(-4, 4, 200)
pdf_normal = norm.pdf(x_normal, mu, sigma)
cdf_normal = norm.cdf(x_normal, mu, sigma)

# Grafica
plt.plot(x_normal, pdf_normal, label="PDF")
plt.plot(x_normal, cdf_normal, label="CDF", linestyle='dashed')
plt.title('Distribución Normal Estándar (μ=0, σ=1)')
plt.xlabel('x')
plt.ylabel('Densidad / Probabilidad')
plt.legend()
plt.show()

# 3. Distribución Exponencial (λ=1)
print("Distribución Exponencial (λ=1):")
lambda_exp = 1
x_expon = np.linspace(0, 8, 200)
pdf_expon = expon.pdf(x_expon, scale=1/lambda_exp)
cdf_expon = expon.cdf(x_expon, scale=1/lambda_exp)

# Grafica
plt.plot(x_expon, pdf_expon, label="PDF")
plt.plot(x_expon, cdf_expon, label="CDF", linestyle='dashed')
plt.title('Distribución Exponencial (λ=1)')
plt.xlabel('x')
plt.ylabel('Densidad / Probabilidad')
plt.legend()
plt.show()
```

**Chapter 5: Normal Distribution**

```python
# Conceptos básicos
print("Distribución Normal y distribuciones relacionadas:\n")

normal_def = "Distribución Normal: una distribución continua simétrica en forma de campana, centrada en su media μ con desviación estándar σ."
z_score_def = "Valor Z: mide cuántas desviaciones estándar un valor x está por encima o por debajo de la media."
chi2_def = "Distribución Chi-cuadrado (χ²): suma de los cuadrados de variables normales estándar independientes."
student_t_def = "Distribución t de Student: distribución usada cuando el tamaño de muestra es pequeño y/o varianza desconocida."
f_dist_def = "Distribución F: compara dos varianzas (relación de dos variables chi-cuadrado independientes)."

print(f"Definición: {normal_def}")
print(f"Valor Z: {z_score_def}")
print(f"Chi-cuadrado: {chi2_def}")
print(f"t de Student: {student_t_def}")
print(f"F de Fisher: {f_dist_def}\n")

# Importar librerías
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, chi2, t, f

# Funciones normales
def z_score(x, mu, sigma):
    return (x - mu) / sigma

def probability_less_than(x, mu, sigma):
    return norm.cdf(x, loc=mu, scale=sigma)

def probability_between(x1, x2, mu, sigma):
    return norm.cdf(x2, loc=mu, scale=sigma) - norm.cdf(x1, loc=mu, scale=sigma)

# Ejemplo de Distribución Normal

print("Ejemplo de Distribución Normal:\n")

mu, sigma = 100, 15
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 200)
y = norm.pdf(x, mu, sigma)

plt.plot(x, y)
plt.title('Distribución Normal (μ=100, σ=15)')
plt.xlabel('x')
plt.ylabel('Densidad de probabilidad')
plt.show()

# Distribuciones relacionadas

# 1. Distribución Chi-cuadrado con 5 grados de libertad
print("Distribución Chi-cuadrado (df=5):\n")

df_chi2 = 5
x_chi2 = np.linspace(0, 20, 200)
y_chi2 = chi2.pdf(x_chi2, df_chi2)

plt.plot(x_chi2, y_chi2, label="Chi² (df=5)")
plt.title('Distribución Chi-cuadrado (df=5)')
plt.xlabel('x')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.show()

# 2. Distribución t de Student con 10 grados de libertad
print("Distribución t de Student (df=10):\n")

df_t = 10
x_t = np.linspace(-5, 5, 200)
y_t = t.pdf(x_t, df_t)

plt.plot(x_t, y_t, label="t (df=10)", color="green")
plt.title('Distribución t de Student (df=10)')
plt.xlabel('x')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.show()

# 3. Distribución F con (5, 2) grados de libertad
print("Distribución F (d1=5, d2=2):\n")

d1, d2 = 5, 2
x_f = np.linspace(0, 5, 200)
y_f = f.pdf(x_f, d1, d2)

plt.plot(x_f, y_f, label="F (5,2)", color="purple")
plt.title('Distribución F (5,2)')
plt.xlabel('x')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.show()

```

We will now examine a third dataset, material strength data collected by Vangel from US NIST.

```python
!wget https://www.itl.nist.gov/div898/software/dataplot/data/VANGEL5.DAT -O VANGEL5.DAT
import pandas as pd
vangel = pd.read_csv("VANGEL5.DAT", header=None, skiprows=25)
vangel = vangel.squeeze("columns")
vangel.head()
```

```python
plt.hist(vangel, density=True, alpha=0.5)
plt.title("Vangel material data (n={})".format(len(vangel)))
plt.xlabel("Specimen strength");
```

Let's try to fit a Weibull distribution to this data.



```python
plt.hist(vangel, density=True, alpha=0.5)
shape, loc, scale = scipy.stats.weibull_min.fit(vangel, floc=0)
support = numpy.linspace(vangel.min(), vangel.max(), 100)
plt.plot(support, scipy.stats.weibull_min(shape, loc, scale).pdf(support), lw=3)
plt.title("Weibull fit on Vangel data")
plt.xlabel("Specimen strength");
```

```python
import statsmodels.distributions

ecdf = statsmodels.distributions.ECDF(vangel)
plt.plot(support, ecdf(support), label="Empirical CDF")
plt.plot(support, scipy.stats.weibull_min(shape,loc,scale).cdf(support), lw=3, label="Weibull fit")
plt.title("Vangel cumulative failure intensity")
plt.xlabel("Specimen strength")
plt.legend();
```

```python
scipy.stats.probplot(vangel, \
                     dist=scipy.stats.weibull_min(shape,loc,scale),\
                     plot=plt.figure().add_subplot(111))
plt.title("Weibull prob-plot of Vangel data");
```

This data follows a Weibull distribution quite well.

If we wish to estimate a particular quantile measure for this data, we can do that on the fitted distribution.

```python
# the first percentile
scipy.stats.weibull_min(shape,loc,scale).ppf(0.01)
```

```python
# Conceptos básicos
print("Estimación estadística y distribuciones muestrales:\n")

point_estimation_def = "Estimación puntual: estimar un parámetro poblacional mediante un solo valor calculado de la muestra."
interval_estimation_def = "Estimación por intervalo: rango de valores con un nivel de confianza donde se espera que esté el parámetro poblacional."
properties_estimators_def = "Propiedades de los estimadores: insesgadez (unbiasedness), eficiencia (efficiency) y consistencia (consistency)."
sampling_distribution_def = "Distribución muestral: distribución de un estadístico basado en todas las muestras posibles de un tamaño dado."

print(f"Estimación puntual: {point_estimation_def}")
print(f"Estimación por intervalo: {interval_estimation_def}")
print(f"Propiedades de los estimadores: {properties_estimators_def}")
print(f"Distribución muestral: {sampling_distribution_def}\n")

# Importar librerías
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Funciones importantes

def confidence_interval_mean(sample, confidence=0.95):
    """Calcula intervalo de confianza para la media con varianza desconocida."""
    n = len(sample)
    mean = np.mean(sample)
    sem = stats.sem(sample)  # error estándar
    h = sem * stats.t.ppf((1 + confidence) / 2, n-1)
    return mean - h, mean + h

def sampling_distribution_plot(n_samples=1000, sample_size=30, population_mean=50, population_std=10):
    """Genera y grafica la distribución muestral de la media."""
    sample_means = []
    for _ in range(n_samples):
        sample = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)
        sample_means.append(np.mean(sample))

    plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='skyblue')
    plt.title(f'Distribución muestral (n={sample_size})')
    plt.xlabel('Media muestral')
    plt.ylabel('Densidad')
    plt.axvline(population_mean, color='red', linestyle='dashed', label=f'Media poblacional ({population_mean})')
    plt.legend()
    plt.show()

# Ejemplo de Intervalo de Confianza

print("Ejemplo de Intervalo de Confianza para la media:\n")

# Crear muestra simulada
np.random.seed(42)
sample_data = np.random.normal(100, 15, 30)

# Calcular intervalo de confianza
ci_lower, ci_upper = confidence_interval_mean(sample_data)
print(f"Intervalo de confianza al 95% para la media: ({ci_lower:.2f}, {ci_upper:.2f})\n")

# Ejemplo de Distribución Muestral

print("Ejemplo de Distribución Muestral:\n")
sampling_distribution_plot(n_samples=1000, sample_size=30, population_mean=100, population_std=15)
```

```python
# Conceptos básicos
print("Inferencias sobre la media poblacional:\n")

hypothesis_testing_def = "Prueba de hipótesis: procedimiento estadístico para decidir entre dos hipótesis sobre un parámetro poblacional."
null_hypothesis_def = "Hipótesis nula (H0): afirmación inicial (sin efecto)."
alternative_hypothesis_def = "Hipótesis alternativa (H1): afirmación que aceptamos si hay evidencia suficiente contra H0."
p_value_def = "Valor p: probabilidad de obtener resultados tan extremos como el observado, asumiendo que H0 es verdadera."
errors_def = "Errores: Tipo I (rechazar H0 siendo verdadera), Tipo II (no rechazar H0 siendo falsa)."
power_def = "Potencia de la prueba: 1 - β (probabilidad de rechazar correctamente H0 cuando H1 es verdadera)."

print(f"Prueba de hipótesis: {hypothesis_testing_def}")
print(f"Hipótesis nula: {null_hypothesis_def}")
print(f"Hipótesis alternativa: {alternative_hypothesis_def}")
print(f"Valor p: {p_value_def}")
print(f"Errores: {errors_def}")
print(f"Potencia de la prueba: {power_def}\n")

# Importar librerías
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Función para prueba t
def one_sample_t_test(sample, mu0, alpha=0.05, alternative='two-sided'):
    """
    Realiza una prueba t para una media poblacional.
    """
    n = len(sample)
    mean_sample = np.mean(sample)
    std_sample = np.std(sample, ddof=1)
    t_stat = (mean_sample - mu0) / (std_sample / np.sqrt(n))
    df = n - 1

    if alternative == 'two-sided':
        p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))
    elif alternative == 'greater':
        p_value = 1 - stats.t.cdf(t_stat, df)
    elif alternative == 'less':
        p_value = stats.t.cdf(t_stat, df)
    else:
        raise ValueError("alternative debe ser 'two-sided', 'greater' o 'less'.")

    reject_null = p_value < alpha
    return t_stat, p_value, reject_null

# Función para calcular error tipo II y potencia
def compute_beta_power(mu0, mu1, sigma, n, alpha=0.05, alternative='two-sided'):
    """
    Calcula la probabilidad de error tipo II (β) y la potencia de la prueba.
    """
    se = sigma / np.sqrt(n)
    df = n - 1

    if alternative == 'two-sided':
        t_critical = stats.t.ppf(1 - alpha/2, df)
        lower_bound = mu0 - t_critical * se
        upper_bound = mu0 + t_critical * se
        beta = stats.norm.cdf((upper_bound - mu1) / se) - stats.norm.cdf((lower_bound - mu1) / se)
    elif alternative == 'greater':
        t_critical = stats.t.ppf(1 - alpha, df)
        beta = stats.norm.cdf((mu0 + t_critical * se - mu1) / se)
    elif alternative == 'less':
        t_critical = stats.t.ppf(alpha, df)
        beta = 1 - stats.norm.cdf((mu0 + t_critical * se - mu1) / se)
    else:
        raise ValueError("alternative debe ser 'two-sided', 'greater' o 'less'.")

    power = 1 - beta
    return beta, power

# Ejemplo de Prueba de Hipótesis

print("Ejemplo de prueba t de una muestra (dos colas):\n")

# Crear muestra simulada
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)

# H0: mu = 105
t_stat, p_value, reject = one_sample_t_test(sample_data, mu0=105)

print(f"Estadístico t = {t_stat:.4f}")
print(f"Valor p = {p_value:.4f}")
print(f"¿Rechazar H0?: {'Sí' if reject else 'No'}\n")

# Ejemplo de Error Tipo II y Potencia

print("Cálculo de Error Tipo II y Potencia:\n")

mu0 = 105       # valor bajo H0
mu1 = 100       # valor verdadero bajo H1
sigma = 15      # desviación estándar conocida
n = 30          # tamaño de muestra
alpha = 0.05    # nivel de significancia

beta, power = compute_beta_power(mu0, mu1, sigma, n, alpha, alternative='two-sided')
print(f"Error Tipo II (β) = {beta:.4f}")
print(f"Potencia de la prueba (1-β) = {power:.4f}\n")

# Gráfica de regiones de rechazo

print("Gráfica de regiones de rechazo:\n")

x = np.linspace(-4, 4, 300)
y = stats.t.pdf(x, df=n-1)

t_critical = stats.t.ppf(1 - alpha/2, df=n-1)

plt.plot(x, y, label='Distribución t (df=29)')
plt.fill_between(x, 0, y, where=(x <= -t_critical) | (x >= t_critical), color='red', alpha=0.3, label='Regiones de rechazo')
plt.axvline(t_stat, color='blue', linestyle='--', label=f'Estadístico t={t_stat:.2f}')
plt.title('Prueba t de dos colas')
plt.xlabel('t')
plt.ylabel('Densidad')
plt.legend()
plt.show()
```

```python
# Useful libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.stats.weightstats as sms
from scipy.stats import t
from scipy.stats import norm
import math
from IPython.display import Math, display
```

Two Sample Problems
Suppose we have:

* Set of data observations
 from a population A with cumulative dist.
* Set of data observations
 from a population B with cumulative dist.
 goal: compare the means

Analysis of Paired Samples

```python
# Parameter
data_A = np.array([23.6, 27.9, 22.9, 21.8, 25.8, 30.7, 26.5, 25.4])
data_B = np.array([22.5, 25.6, 24.0, 20.4, 26.0, 26.6, 26.4, 22.1])
z = data_A - data_B

# Mean
z_bar = z.mean(); s = z.std(ddof=1) # Same as pandas
print('Mean = {:.4f}\nStandard deviation = {:.4f}'.format(z_bar, s))

""" We test
$H_0 : \mu \leq 0$
$H_A : \mu > 0$ """
n = len(z)
mu = 0
t_stat = math.sqrt(n)*(z_bar - mu)/s
print('t-statistics: {:.4f}'.format(t_stat))
print(("p-value = {:.4f}".format(t.sf(t_stat, n-1))))
```

Analysis of Independent Samples


```python
# Population A
n = 14; x_bar = 32.45; s_x = 4.30

# Population B
m = 14; y_bar = 41.45; s_y = 5.23

alpha = 0.01 # = 1 - confidence level

def degrees_of_freedom(n, m, s_x, s_y):
    nu =  ((s_x**2/n + s_y**2/m)**2) / ( s_x**4 / (n**2*(n-1)) + s_y**4 / (m**2*(m-1)))
    return math.floor(nu)

def wing_span(alpha, nu, n, m, s_x, s_y):
    return t.ppf(1-alpha/2, nu) * math.sqrt( s_x**2/n + s_y**2/m)

def print_confidence_interval(alpha, mu, wing_span):
    print(("Confidence interval with {:.2f}% confidence level: ({:.4f})({:.4f})").format(((1-alpha)*100), mu - wing_span, mu + wing_span))

# Calculations
diff = x_bar - y_bar
nu = degrees_of_freedom( n, m, s_x, s_y)
wing_span = wing_span(alpha, nu, n, m, s_x, s_y)
display(Math('v = {}'.format(nu)))
print_confidence_interval(alpha, diff, wing_span)
```

```python
"""Hypotheses:
- $H_0 : \mu_A = \mu_B$
- $H_A : \mu_A \neq \mu_B$"""

# Calculate the t-statistics general
def T_statistic_general(x_bar, y_bar, s_x, s_y, n, m, delta= 0):
    return ( (x_bar - y_bar - delta)  / math.sqrt(s_x**2/n + s_y**2/m))

t_stat = T_statistic_general(x_bar, y_bar, s_x, s_y, n, m, delta= 0)
print("|t| value: {:.4f} ".format(abs(t_stat)))
print("Critical point: {:.4f}".format(t.ppf(1- alpha/2, n + m -2 )))
```

Pooled Variance Procedure


```python
# Population A
n = 14; x_bar = 32.45; s_x = 4.30

# Population B
m = 14; y_bar = 41.45; s_y = 5.23

alpha = 0.01 # = 1 - confidence level

# Pooled variance
def pooled_variance(n, m, s_x, s_y):
    return ( ((n-1)*s_x**2 + (m-1)*s_y**2) / (n+m-2))

def wing_span(s_p, n, m, alpha):
    return t.ppf(1-alpha/2, n+m-2) * s_p*math.sqrt(1/n + 1/m)

def print_confidence_interval(alpha, mu, wing_span):
    print(("Confidence interval with {:.2f}% confidence level: ({:.4f})({:.4f})").format(((1-alpha)*100), mu - wing_span, mu + wing_span))

diff = x_bar - y_bar
s_p = math.sqrt(pooled_variance(n, m, s_x, s_y))
wing_span = wing_span(s_p, n, m, alpha)
print_confidence_interval(alpha, diff, wing_span)
```

```python
"""Hypotheses:
- $H_0 : \mu_A = \mu_B$
- $H_A : \mu_A \neq \mu_B$"""

# Calculate the t-statistics
def T_statistic_pooled(x_bar, y_bar, s_p, n, m, delta= 0):
    return ( (x_bar - y_bar - delta)  / (s_p * math.sqrt(1/n + 1/m)))

t_stat = T_statistic_pooled(x_bar, y_bar, s_p, n, m, delta= 0)
print("|t| value: {:.4f} ".format(abs(t_stat)))
print("Critical point: {:.4f}".format(t.ppf(1- alpha/2, n + m -2 )))
```

z-Procedure


```python
# Population A
n = 38; x_bar = 5.782; sigma_x= 2.0

# Population B
m = 40; y_bar = 6.443; sigma_y= 2.0

alpha = 0.01 # = 1 - confidence level

# Calculate the Z-statistics
def Z_statistic(x_bar, y_bar, s_x, s_y, n, m, delta = 0):
    return ( (x_bar - y_bar - delta)  / math.sqrt(s_x**2/n + s_y**2/m))

Z_stat = Z_statistic(x_bar, y_bar, sigma_x, sigma_y, n, m)
print("Z-statistics: {:.4f}".format(Z_stat))
print("p-value: {:.4f} ".format(norm.cdf(Z_stat)))

def wing_span(alpha, n, m, s_x, s_y):
    return norm.ppf(1-alpha) * math.sqrt( s_x**2/n + s_y**2/m)

def print_confidence_interval(alpha, mu, wing_span, lower_bound = False, upper_bound = False):
    if upper_bound:
        print(("Confidence interval with {:.2f}% confidence level: (-∞)({:.4f})").format(((1-alpha)*100), mu + wing_span))
        return
    if lower_bound:
        print(("Confidence interval with {:.2f}% confidence level: ({:.4f})(∞))").format(((1-alpha)*100), mu - wing_span))
        return
    # Default case: double bounded
    print(("Confidence interval with {:.2f}% confidence level: ({:.4f})({:.4f})").format(((1-alpha)*1, mu - wing_span, mu + wing_span)))

"""Hypotheses:
- $H_0 : \mu_A > \mu_B$''
- $H_A : \mu_A >= \mu_B$"""
diff = x_bar - y_bar
wing_span = wing_span(alpha, n, m, sigma_x, sigma_y)
print_confidence_interval(alpha, diff, wing_span, upper_bound = True)
```

**Chapter 11: Analysis of Variance**

```python
# Definiciones teóricas
print("Análisis de Varianza (ANOVA):\n")

anova_def = "ANOVA es una técnica estadística para comparar las medias de tres o más grupos mediante el análisis de la variabilidad."
hypotheses_def = "H0: Todas las medias poblacionales son iguales. H1: Al menos una media es diferente."
f_stat_def = "Estadístico F: cociente entre la varianza entre grupos y la varianza dentro de los grupos."
assumptions_def = "Supuestos: independencia de muestras, normalidad en los grupos, y varianzas homogéneas (homocedasticidad)."

print(f"Definición: {anova_def}")
print(f"Hipótesis: {hypotheses_def}")
print(f"Estadístico F: {f_stat_def}")
print(f"Supuestos: {assumptions_def}\n")

# Importar librerías
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Ejemplo práctico con 3 grupos

print("Ejemplo práctico de ANOVA:\n")

np.random.seed(42)
group1 = np.random.normal(loc=50, scale=5, size=30)
group2 = np.random.normal(loc=52, scale=5, size=30)
group3 = np.random.normal(loc=60, scale=5, size=30)

# Realizar ANOVA de un solo factor
f_stat, p_value = stats.f_oneway(group1, group2, group3)

print(f"Estadístico F: {f_stat:.4f}")
print(f"Valor p: {p_value:.4f}")
print(f"¿Rechazar H0?: {'Sí' if p_value < 0.05 else 'No'}\n")

# Visualización con Seaborn

# DataFrame
data = pd.DataFrame({
    'valor': np.concatenate([group1, group2, group3]),
    'grupo': ['Grupo 1']*30 + ['Grupo 2']*30 + ['Grupo 3']*30
})

# Boxplot
sns.boxplot(x='grupo', y='valor', data=data)
plt.title('Comparación de grupos (Boxplot)')
plt.ylabel('Valor observado')
plt.show()

# Violin plot
sns.violinplot(x='grupo', y='valor', data=data)
plt.title('Comparación de grupos (Violin Plot)')
plt.ylabel('Distribución estimada')
plt.show()

# Interpretación

print("Interpretación de resultados:")
if p_value < 0.05:
    print("Se rechaza H0: existe evidencia significativa de que al menos una media es diferente.")
else:
    print("No se puede rechazar H0: no hay evidencia suficiente para afirmar que las medias difieren.\n")
```

**Chapter 12: Simple Linear Regression and Correlation**

```python

print("🔹 Regresión Lineal Simple:")
print("Modelo que describe la relación entre una variable independiente (X) y una dependiente (Y) mediante una recta.")
print("Ecuación: Y = a + bX, donde:")
print("- a: intercepto (valor de Y cuando X = 0)")
print("- b: pendiente (cambio en Y por unidad de X)\n")

print("🔹 Objetivo:")
print("Estimar y predecir el valor de Y para un valor dado de X con base en datos observados.\n")

print("🔹 Coeficiente de Correlación (r):")
print("Mide la fuerza y dirección de la relación lineal entre X e Y.")
print("Rango: [-1, 1]")
print("- r > 0: relación positiva")
print("- r < 0: relación negativa")
print("- r = 0: sin relación lineal\n")

print("🔹 Coeficiente de Determinación (R²):")
print("Proporción de la variabilidad en Y explicada por X.")
print("R² = (r)² y varía entre 0 y 1.\n")

print("🔹 Supuestos del modelo:")
print("- Relación lineal entre X e Y")
print("- Residuos con media cero")
print("- Varianza constante de los residuos (homocedasticidad)")
print("- Normalidad de los residuos")
print("- Observaciones independientes\n")

print("🔹 Predicción:")
print("La regresión permite estimar un valor de Y para un nuevo valor X y calcular intervalos de confianza y predicción.\n")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Datos simulados
np.random.seed(0)
X = np.random.rand(50) * 10
Y = 2.5 * X + np.random.normal(0, 2, size=X.shape)

# Ajustar modelo
X_reshape = X.reshape(-1, 1)
model = LinearRegression()
model.fit(X_reshape, Y)
Y_pred = model.predict(X_reshape)

# Parámetros del modelo
slope = model.coef_[0]
intercept = model.intercept_
r_value, _ = stats.pearsonr(X, Y)
r2 = r2_score(Y, Y_pred)

print("📈 Resultados:")
print(f"Pendiente: {slope:.4f}")
print(f"Intercepto: {intercept:.4f}")
print(f"r (correlación): {r_value:.4f}")
print(f"R² (determinación): {r2:.4f}\n")

# ---------- Intervalo de confianza y predicción ----------
nuevo_x = 6.5
n = len(X)
mean_x = np.mean(X)
se = np.sqrt(np.sum((Y - Y_pred)**2) / (n - 2))
t_crit = stats.t.ppf(0.975, df=n-2)

# Intervalo de confianza del promedio
conf_mean = model.predict([[nuevo_x]])[0]
margin_mean = t_crit * se * np.sqrt(1/n + ((nuevo_x - mean_x)**2 / np.sum((X - mean_x)**2)))
ci_lower = conf_mean - margin_mean
ci_upper = conf_mean + margin_mean

# Intervalo de predicción individual
margin_pred = t_crit * se * np.sqrt(1 + 1/n + ((nuevo_x - mean_x)**2 / np.sum((X - mean_x)**2)))
pi_lower = conf_mean - margin_pred
pi_upper = conf_mean + margin_pred

print(f"🔍 Predicción para X = {nuevo_x}: Y = {conf_mean:.2f}")
print(f"IC 95% para la media: ({ci_lower:.2f}, {ci_upper:.2f})")
print(f"IP 95% para un valor individual: ({pi_lower:.2f}, {pi_upper:.2f})\n")

# ---------- Gráfica con IC y residuos ----------
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X, y=Y, label="Datos reales")
plt.plot(X, Y_pred, color='red', label='Recta de regresión')
plt.axhline(y=conf_mean, color='green', linestyle='--', label='Predicción media')
plt.fill_between([nuevo_x], ci_lower, ci_upper, color='green', alpha=0.3, label='IC 95%')
plt.fill_between([nuevo_x], pi_lower, pi_upper, color='orange', alpha=0.3, label='IP 95%')
plt.scatter(nuevo_x, conf_mean, color='black', zorder=5)
plt.title("Regresión con intervalos de confianza y predicción")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()

# ---------- Análisis de residuos ----------
residuos = Y - Y_pred

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.residplot(x=Y_pred, y=residuos, lowess=True, line_kws={'color': 'red'})
plt.axhline(0, color='gray', linestyle='--')
plt.title("Residuos vs. Valores ajustados")
plt.xlabel("Valores ajustados")
plt.ylabel("Residuos")

plt.subplot(1,2,2)
sns.histplot(residuos, kde=True, bins=10)
plt.title("Distribución de residuos")
plt.xlabel("Residuos")
plt.tight_layout()
plt.show()
```
